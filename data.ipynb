# Accenture Data Analysis for Social Buzz

### Import Necessary Libraries

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

### Load Necessary Files:

#### 1. Load Content File:

```python
Content_df= pd.read_csv("Content.csv")
Content_df
```

#### Cleaning and Transform the Data:

```python
Content_df = Content_df.drop(columns=["URL", "Unnamed: 0"])
Content_df
```

```python
Content_df['Category']=Content_df['Category'].str.replace('"','')
Content_df['Category']=Content_df['Category'].str.lower()
Content_df
```

```python
df= Content_df["Category"].isnull()
df
```

```python
Content_df["Content Type"] = Content_df["Type"]
Content_df = Content_df.drop(columns="Type")
Content_df
```

#### 2. Load Reactions File:

```python
Reactions_df= pd.read_csv("Reactions.csv")
Reactions_df
```

#### Cleaning and Transform the Data:

```python
Reactions_df=Reactions_df.drop(columns="Unnamed: 0")
Reactions_df
```

```python
Reactions_df=Reactions_df.drop_duplicates()
Reactions_df
```

```python
Reactions_df["reactionnull"] = Reactions_df["Type"].isnull()
Reactions_df["usernull"]= Reactions_df["User ID"].isnull()
Reactions_df
```

```python
Reactions_df=Reactions_df.loc[(Reactions_df["reactionnull"]==False) & (Reactions_df["usernull"]==False)]
Reactions_df
```

```python
Reactions_df=Reactions_df.drop(columns=["usernull","reactionnull"])
Reactions_df
```

#### 3.Load Reaction Types file:

```python
Reactiontype_df= pd.read_csv("ReactionTypes.csv")
Reactiontype_df
```

#### Cleaning and Transform the Data:

```python
Reactiontype_df=Reactiontype_df.drop(columns="Unnamed: 0")
Reactiontype_df=Reactiontype_df.drop_duplicates()
Reactiontype_df
```

### Join The Files:

```python
First_df=Content_df.merge(Reactions_df,on="Content ID",how="inner")
First_df
```

```python
FinalData_df=First_df.merge(Reactiontype_df,on="Type",how="inner")
FinalData_df
```

#### Cleaning the Joined Data:

```python
FinalData_df=FinalData_df.drop(columns=["User ID_x","User ID_y"])
FinalData_df
```

### Export Final Data Frame into CSV:

```python
FinalData_df.to_csv("FinalData.csv", index=False)
```

### Load Final Data File:

```python
FinalData_df
```

#### Extract and Visualize Top Content Type According to the Score:

```python
ndf = FinalData_df.groupby("Content Type").agg(Max_score= ("Score", sum))
ndf =ndf.reset_index()
ndf.sort_values(by="Max_score", ascending=False)
```

```python
plt.bar(ndf['Content Type'],ndf['Max_score'],width=0.5)
plt.title('Content_Type by Score')
plt.xlabel('Content Type')
plt.ylabel('Score')
plt.grid()
plt.show()
```

```python
plt.pie(ndf["Max_score"], labels= ndf["Content Type"], autopct='%.2f%%', explode=[0,0,0,0.1], radius=1.5)
plt.show()
```

#### Extract and Visualize Top 5 Category According to the Score:

```python
cdf=FinalData_df.groupby("Category").agg(Max_score= ("Score", sum))
cdf =cdf.reset_index()
cdf=cdf.sort_values(by="Max_score", ascending=False)
cdf=cdf.head(5)
cdf
```

```python
cdf.to_csv("Top5categories.csv", index=False)
```

```python
plt.plot(cdf['Category'],cdf['Max_score'],linewidth=2,marker='o')
plt.title('Category by Score')
plt.xlabel('Category')
plt.ylabel('Score')
plt.grid()
plt.show()
```

#### Extract Best Months According to the Score:

```python
FinalData_df['Datetime']=pd.to_datetime(FinalData_df['Datetime'])
FinalData_df
```

```python
FinalData_df["Month"]=FinalData_df['Datetime'].dt.month_name()
FinalData_df
```

```python
cdf_1=FinalData_df.groupby("Month").agg(Max_score= ("Score", sum))
cdf_1 =cdf_1.reset_index()
cdf_1=cdf_1.sort_values(by="Max_score", ascending=False)
cdf_1
```

```python
plt.plot(cdf_1['Month'],cdf_1['Max_score'],linewidth=2,marker='o')
plt.title('Month wise Views by Score')
plt.xlabel('Month')
plt.ylabel('Score')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

#### Analyse the Content According to the Best Months & Score:

```python
hdf = FinalData_df.groupby(["Content Type", "Month"]).agg(Max_score=('Score','sum'))
hdf = hdf.reset_index()
hdf = hdf.sort_values(by="Max_score", ascending=False)
hdf
```

```python
import seaborn as sns
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.barplot(x="Content Type", y="Max_score", hue="Month", data=hdf)
plt.xticks(rotation=45)
plt.title('Max Score by Content Type and Month')
plt.xlabel('Content Type')
plt.ylabel('Max Score')
plt.legend(title='Month')
plt.show()
```
